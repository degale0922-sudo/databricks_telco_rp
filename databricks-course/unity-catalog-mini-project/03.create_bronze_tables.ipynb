###Crear Tablas en el Schema Bronze
1. clientes.csv desde "abfss://bronze@sadatabricksextdegale.dfs.core.windows.net/clientes.csv"
2. detalles.csv desde "abfss://bronze@sadatabricksextdegale.dfs.core.windows.net/detalles.csv"

```sql
drop table if exists telco_dev.bronze.clientes;
create table if not exists telco_dev.bronze.clientes(
  clienteId int,
  nombre string,
  paterno string,
  materno string,
  documentoIdentidad string,
  telefono long,
  email string,
  direccion string
)
using csv
options (path "abfss://bronze@sadatabricksextdegale.dfs.core.windows.net/clientes.csv", delimiter ";", header true)
```

```sql
select * from telco_dev.bronze.clientes
```

```sql
drop table if exists telco_dev.bronze.detalles;
create table if not exists telco_dev.bronze.detalles(
  detalleId int,
  clienteId int,
  servicio string,
  plan string,
  precioMensual decimal(5,2),
  estado boolean,
  fechaInicio date,
  fechaFin date
)
using csv
options (path "abfss://bronze@sadatabricksextdegale.dfs.core.windows.net/detalles.csv", delimiter ";", header true)
```

```sql
select * from telco_dev.bronze.detalles
```
