###Crear Managed Table en el Schema gold
- Usar JOIN en las tablas "clientes" y "detalles" para obtener las "ingresos por clientes"

```sql
drop table if exists telco_dev.gold.ingreso_por_clientes;
create table telco_dev.gold.ingreso_por_clientes
as 
  select C.nombre_completo, SUM(D.precio_mensual) as ingreso
  from telco_dev.silver.clientes C
  inner join telco_dev.silver.detalles D
  on C.cliente_id = D.cliente_id
  where D.estado = true
  group by C.nombre_completo
  order by ingreso desc
```

```sql
select * from telco_dev.gold.ingreso_por_clientes;
```
